# Scraping the data and links of all the states
1. Target website =  http://www.city-data.com
2. Target class: cities_list

```python
options = Options()
#driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=options) 
driver = webdriver.Chrome()
driver.maximize_window()
url = 'http://www.city-data.com'
driver.get(url)
time.sleep(5)
states = driver.find_elements(By.CLASS_NAME,'cities_list')

state_links = []
state_names = []
for i in states:
    state_links.append(i.get_attribute('href'))
    state_names.append(i.text)
    
driver.close()
```

```python
state_links = state_links[52:]
state_names = state_names[52:]

for i in range (0,52):
    print(state_names[i]," : ",state_links[i])
```

# Data sets
Data sets should containg the following information about all of the states. 
1. Capital 
2. 2019 population 
3. Change from 2000 
4. Births 2000-2019
5. Deaths 2000-2019
6. Internal migration 2000-2019
7. International migration 2000-2019
8. 2019 housing units
9. Housing units change from 2000
10. Land area
11. State nickname

```python
stateName, DS1_a0 ,DS1_a1 ,DS1_a2,DS1_a3,DS1_a4,DS1_a5,DS1_a6,DS1_a7,DS1_a8,DS1_a9,DS1_a10 = [],[],[],[],[],[],[],[],[],[],[], []
```

```python
options = Options()
#driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=options) 
driver = webdriver.Chrome()
driver.maximize_window()

leftover_index = []
leftover_link =[]
for state in range (0, len(state_links)):
    try:
        driver.get(state_links[state])
    
        #Getting the values of the 11 fields
        list_group = driver.find_element(By.CLASS_NAME,'list-group')
        values = list_group.find_elements(By.CLASS_NAME,'badge')
    
        if(len(values) == 11):
            DS1_a0.append(values[0].text)
            DS1_a1.append(values[1].text)
            DS1_a2.append(values[2].text)
            DS1_a3.append(values[3].text)
            DS1_a4.append(values[4].text)
            DS1_a5.append(values[5].text)
            DS1_a6.append(values[6].text)
            DS1_a7.append(values[7].text)
            DS1_a8.append(values[8].text)
            DS1_a9.append(values[9].text)
            DS1_a10.append(values[10].text)
            stateName.append(state_names[state])
        
        else:
            leftover_link.append(state_links[state])
            leftover_index.append(state)
            print("We have a value exception at state link index ",state)
    
    except:
        print("State information not available at state link index ",state)
        continue
        
driver.quit()
```

```python
data = {
    "State":stateName,
    "Capital":DS1_a0,
    "2019 population":DS1_a1,
    "Change from 2000":DS1_a2,
    "Births 2000-2019":DS1_a3,
    "Deaths 2000-2019":DS1_a4,
    "Internal migration 2000-2019":DS1_a5,
    "International migration 2000-2019":DS1_a6,
    "2019 housing units":DS1_a7,
    "Housing units change from 2000":DS1_a8,
    "Land area":DS1_a9,
    "State nickname":DS1_a10,
}

# Create the DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```
